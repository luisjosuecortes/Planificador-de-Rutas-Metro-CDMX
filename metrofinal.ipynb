```python
# Definir la clase TGen
class TGen:
    def __init__(self, linea, estacion):
        self.linea = linea  # entero: posición en la lista de líneas del plano
        self.estacion = estacion  # entero: posición de la estación en la línea

# Función para obtener el nombre de la estación
def obtener_nombre_estacion(tgen, lineas):
    """Obtiene el nombre de la estación a partir de un objeto TGen."""
    linea = tgen.linea
    estacion = tgen.estacion
    if linea in lineas and 0 <= estacion < len(lineas[linea]):
        return lineas[linea][estacion]
    else:
        return None

# Preprocesar las estaciones para mapear nombre a línea y posición
station_map = defaultdict(dict)
for linea_num, estaciones in lineas.items():
    for idx, estacion in enumerate(estaciones):
        station_map[estacion][linea_num] = idx

# Clase Nodo para A*
class Nodo:
    def __init__(self, estado, padre=None, stations=0, transfers=0, heuristica=0, required_collected=None):
        self.estado = estado
        self.padre = padre
        self.stations = stations  # número de estaciones recorridas
        self.transfers = transfers  # número de transbordos
        self.heuristica = heuristica
        self.required_collected = required_collected if required_collected is not None else set()

    def __lt__(self, otro):
        # Prioridad: (estaciones + heurística) * 1000 + transbordos
        total_self = (self.stations + self.heuristica) * 1000 + self.transfers
        total_otro = (otro.stations + otro.heuristica) * 1000 + otro.transfers
        return total_self < total_otro

# Heurística para A* (distancia de Manhattan modificada)
def heuristica(estado_actual, estado_final):
    return abs(estado_actual.linea - estado_final.linea) + abs(estado_actual.estacion - estado_final.estacion)

# Obtener el camino desde el nodo inicial hasta el final
def obtener_camino(nodo):
    camino = []
    while nodo:
        camino.append(nodo.estado)
        nodo = nodo.padre
    return camino[::-1]

# Implementación del algoritmo A* mejorado
def a_estrella(estacion_inicial, estacion_final, conexiones, lineas, est_prohibidas, est_obligadas):
    frontera = []
    initial_name = obtener_nombre_estacion(estacion_inicial, lineas)
    initial_required = set([initial_name]) if initial_name in est_obligadas else set()
    heapq.heappush(
        frontera,
        Nodo(
            estacion_inicial,
            None,
            0,
            0,
            heuristica(estacion_inicial, estacion_final),
            initial_required
        )
    )
    explorados = set()

    while frontera:
        nodo_actual = heapq.heappop(frontera)
        estado_actual = nodo_actual.estado
        current_line = estado_actual.linea
        current_pos = estado_actual.estacion
        current_name = obtener_nombre_estacion(estado_actual, lineas)
        current_required = nodo_actual.required_collected

        # Verificar si el estado actual ya fue explorado con los mismos requerimientos
        state_key = (current_line, current_pos, frozenset(current_required))
        if state_key in explorados:
            continue
        explorados.add(state_key)

        # Comprobar si es el estado objetivo y cumple restricciones
        if (estado_actual.linea == estacion_final.linea and
            estado_actual.estacion == estacion_final.estacion and
            current_required.issuperset(est_obligadas)):
            return obtener_camino(nodo_actual)

        # Generar movimientos en la misma línea
        for delta in [1, -1]:
            new_pos = current_pos + delta
            if 0 <= new_pos < len(lineas[current_line]):
                new_estado = TGen(current_line, new_pos)
                new_name = obtener_nombre_estacion(new_estado, lineas)
                if new_name in est_prohibidas:
                    continue
                new_required = set(current_required)
                if new_name in est_obligadas:
                    new_required.add(new_name)
                if (new_estado.linea, new_estado.estacion, frozenset(new_required)) in explorados:
                    continue
                h = heuristica(new_estado, estacion_final)
                new_node = Nodo(
                    new_estado,
                    nodo_actual,
                    nodo_actual.stations + 1,
                    nodo_actual.transfers,
                    h,
                    new_required
                )
                heapq.heappush(frontera, new_node)

        # Generar transbordos
        if current_name in conexiones.get(current_line, {}):
            for nueva_linea in conexiones[current_line][current_name]:
                if nueva_linea not in station_map[current_name]:
                    continue
                new_pos = station_map[current_name][nueva_linea]
                new_estado = TGen(nueva_linea, new_pos)
                new_name = obtener_nombre_estacion(new_estado, lineas)
                if new_name in est_prohibidas:
                    continue
                new_required = set(current_required)
                if new_name in est_obligadas:
                    new_required.add(new_name)
                if (new_estado.linea, new_estado.estacion, frozenset(new_required)) in explorados:
                    continue
                h = heuristica(new_estado, estacion_final)
                new_node = Nodo(
                    new_estado,
                    nodo_actual,
                    nodo_actual.stations,
                    nodo_actual.transfers + 1,
                    h,
                    new_required
                )
                heapq.heappush(frontera, new_node)

    return None
```

```python
# Ejemplo de uso:
estacion_inicial = TGen(1, 0) # Observatorio
estacion_final = TGen(12, 9) 
est_prohibidas = []
est_obligadas = []

camino = a_estrella(estacion_inicial, estacion_final, conexiones, lineas, est_prohibidas, est_obligadas)

if camino:
    print("Camino encontrado:")
    transbordos = 0
    linea_actual = None
    for gen in camino:
        nombre = obtener_nombre_estacion(gen, lineas)
        if linea_actual != gen.linea:
            if linea_actual is not None:
                transbordos += 1
            linea_actual = gen.linea
        print(f"Línea {gen.linea}: {nombre}")
    print(f"Número de transbordos: {transbordos}")
else:
    print("No se encontró un camino que cumpla las restricciones.")
```
